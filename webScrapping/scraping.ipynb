```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
page = requests.get('https://www.flipkart.com/laptops/~buyback-guarantee-on-laptops-/pr?sid=6bo%2Cb5g&uniqBStoreParam1=val1&wid=11.productCard.PMU_V2')
print(page)
```

```python
# Parsing the HTML
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
# Getting the title tag
print(soup.title)
 
# Getting the name of the tag
print(soup.title.name)
 
# Getting the name of parent tag
print(soup.title.parent.name)
```

```python
laptops = soup.find('div',class_="_13oc-S")
```

```python
laptops[0]
```

```python
laptops[1]
```

```python
# the price class="_30jeq3 _1_WHN1"
price = []
name = []
rating = []

prices = soup.find_all('div', class_='_30jeq3 _1_WHN1')
for p in prices:
    price.append(p.text)
    
    
namelist = soup.find_all('div', class_='_4rR01T')
for n in namelist:
    name.append(n.text)
    
ratelist = soup.find_all('div', class_='_3LWZlK')
for n in ratelist:
    rating.append(n.text)
    
print(price)
print(name)
print(rating)
```

```python
df = pd.DataFrame({'Laptop Name':name,'Price':price,'Rating':rating}) 
df.to_csv('products.csv', index=False, encoding='utf-8')
```
